```python
from sklearn.datasets import fetch_20newsgroups
#cats = ['alt.atheism', 'sci.space']
#n_train = fetch_20newsgroups(subset='train', categories=cats)
#list(n_train.target_names)
n_train = fetch_20newsgroups(subset='train')
n_test = fetch_20newsgroups(subset='test')
```

```python
n_train.data[0]
```

```python
n_train.target[0]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
ngram_range = (1,2)
my_tfidf_vec = TfidfVectorizer(input='content', ngram_range=ngram_range)
X_train_transformed = my_tfidf_vec.fit_transform(n_train.data)
```

```python
X_train_transformed
```

```python
from xgboost import XGBClassifier
clf = XGBClassifier()
clf.fit(X_train_transformed, n_train.target)
```

```python
clf.score(X_train_transformed, n_train.target)
```

```python
X_test_transformed = my_tfidf_vec.transform(n_test.data, )
```
